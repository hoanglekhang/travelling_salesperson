INPUT: Matrix of distances between each pair of n cities  
OUTPUT: Best TSP solution generated by the evolutionary algorithm  

INITIALIZE current tour randomly  
CALCULATE distance of the current tour  
SET best tour as current tour  
SET best distance as distance of the current tour  
SET stagnation count to 0  

WHILE stagnation count < MAX_STAGNATIONS:  
    MUTATE current tour to create a new tour  
    CALCULATE distance of the new tour
    
    IF distance of new tour < best distance:
        SET best tour as new tour
        SET best distance as distance of new tour
        RESET stagnation count to 0
    ELSE:
        INCREMENT stagnation count by 1
    
    IF stagnation count == MAX_STAGNATIONS:
        PERFORM three mutations on best tour to escape local optimum
        RESET stagnation count to 0

PRINT best tour and its total distance
